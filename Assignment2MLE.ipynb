# Importing Dependencies

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

# Defining Likelihood and Log-Likelihood Functions

## Likelihood Function

```python
def exponential_distribution_lf(lambda_v, x_vals):
    return (lambda_v ** len(x_vals)) * ( math.e ** (-lambda_v*np.sum(x_vals)))
```

## Log-Likelihood Function

```python
def exponential_distribution_llf(lambda_v, x_vals):
    return len(x_vals) * math.log(lambda_v) - lambda_v * np.sum(x_vals)
```

# Generating Data

```python
true_lambda = 2  # True lambda value for generating data
sample_size = 100  # Number of samples
sample_data = np.random.exponential(1 / true_lambda, sample_size) 
```

```python
lambda_mle = len(sample_data) / np.sum(sample_data)
```

# Computing the likelihood and log-likelihood for the values

```python
lambda_values = np.linspace(0.1, 5, 100)  # Range of lambda values for visualization
likelihoods = [exponential_distribution_lf(l, sample_data) for l in lambda_values]
log_likelihoods = [exponential_distribution_llf(l, sample_data) for l in lambda_values]
```

# Plotting the results

## Log likelihood

```python

plt.figure(figsize=(8, 5))
plt.plot(lambda_values, log_likelihoods, label="Log-Likelihood", color='blue')
plt.axvline(lambda_mle, color='red', linestyle="--", label=f"MLE λ = {lambda_mle:.3f}")
plt.xlabel("Lambda (λ)")
plt.ylabel("Log-Likelihood")
plt.title("Log-Likelihood Function of Exponential Distribution")
plt.legend()
plt.show()
```

## Likelihood

```python
plt.figure(figsize=(8, 5))
plt.plot(lambda_values, likelihoods, label="Likelihood", color='blue')
plt.axvline(lambda_mle, color='red', linestyle="--", label=f"MLE λ = {lambda_mle:.3f}")
plt.xlabel("Lambda (λ)")
plt.ylabel("Likelihood")
plt.title("Likelihood Function of Exponential Distribution")
plt.legend()
plt.show()
```

# Result

```python

# Output results
print(f"Sample Data (First 10 values): {sample_data[:10]}")
print(f"MLE of Lambda: {lambda_mle:.4f}")
print(f"Log-Likelihood at MLE: {exponential_distribution_llf(lambda_mle, sample_data):.4f}")
```
